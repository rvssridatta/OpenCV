```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

#rect3.JPG is an image drawn in ms paint

```python
# load image
img = cv2.imread('rect5.JPG')
# img_r = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(img)
```

```python


##########################################
# At first vertical devide image         #
##########################################
# start vertical devide image
height = img.shape[0]
width = img.shape[1]
# Cut the image in half
width_cutoff = width // 2
left1 = img[:, :width_cutoff]
right1 = img[:, width_cutoff:]
# finish vertical devide image

# c1 = img[:,:width_cutoff]
cv2_imshow(left1)
```

```python
cv2_imshow(right1)
```

```python
width//4
```

```python
quandrant1 = img[:height//2,width//2:]
cv2_imshow(quandrant1)
```

# to find 5th rectangle from midpoint
#consider quandrant3(q3) portion in image quandrant1(Q1)

```python
hi = quandrant1.shape[0]
wi = quandrant1.shape[1]
q1_quandrant3 = quandrant1[hi//2:,:wi//2]
cv2_imshow(q1_quandrant3)
```

```python
quandrant2 = img[:height//2,:width//2]
cv2_imshow(quandrant2)
```

# to find 5th rectangle from midpoint
#consider quandrant4(q4) portion in image quandrant2(Q2)

```python
hi = quandrant2.shape[0]
wi = quandrant2.shape[1]
q2_quandrant4 = quandrant2[hi//2:,wi//2:]
cv2_imshow(q2_quandrant4)
```

```python
quandrant3 = img[height//2:,:width//2]
cv2_imshow(quandrant3)
```

# to find 5th rectangle from midpoint
#consider quandrant1(q1) portion in image quandrant3(Q3)

```python
hi = quandrant3.shape[0]
wi = quandrant3.shape[1]
q3_quandrant1 = quandrant3[:hi//2,wi//2:] #[:height//2,width//2:]
cv2_imshow(q3_quandrant1)
```

```python
quandrant4 = img[height//2:,width//2:]
cv2_imshow(quandrant4)
```

# to find 5th rectangle from midpoint
#consider quandrant2(q2) portion in image quandrant4(Q4)

```python
hi = quandrant4.shape[0]
wi = quandrant4.shape[1]
q4_quandrant2 = quandrant4[:hi//2,:wi//2]
cv2_imshow(q4_quandrant2)
```

merging those parts to get mid rectangle 

```python
vis1 = np.concatenate((q2_quandrant4,q1_quandrant3), axis=1)
cv2_imshow(vis1)
```

```python
vis2 = np.concatenate((q3_quandrant1,q4_quandrant2), axis = 1)
cv2_imshow(vis2)
```

"mid reactangel image"

```python
vis = np.concatenate((vis1,vis2), axis = 0)
cv2_imshow(vis)
```
